```python
import pandas as pd
import numpy as np
```

```python
events = pd.read_csv('course-217-events.csv')
struct = pd.read_csv('course-217-structure.csv')
```

```python
struct = struct.sort_values(['module_position', 'lesson_position', 'step_position'])
struct['abs_step_position'] = np.arange(struct.step_id.unique().shape[0])
```

```python
struct = struct.set_index('step_id')
events['abs_step_position'] = struct.loc[events.step_id].abs_step_position.values
struct = struct.reset_index()
```

```python
views = np.zeros([events.user_id.unique().shape[0], struct.abs_step_position.unique().shape[0]])
views_next = np.zeros_like(views)
returns = np.zeros_like(views)
```

```python
events = events.sort_values('time')
events.user_id -= 1
```

```python
for x in events.values:
    if (x[0] > 0) and (views[x[0], x[4] - 1] == 1.0):
        views_next[x[0], x[4]] = 1.0
    if (views[x[0], x[4]] == 1.0) and (x[4] < 232) and (views_next[x[0], x[4] + 1] == 1.0):
        returns[x[0], x[4]] = 1.0
    views[x[0], x[4]] = 1.0
```

```python
struct = struct.sort_values('abs_step_position')
struct['reab'] = returns.sum(axis = 0) / views.sum(axis = 0)
```
